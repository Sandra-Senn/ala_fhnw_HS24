## Quadratische Formen und Geometrie

```python
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x, y = sp.symbols("x y")
```

Wir benutzen die Funktion `plot_implicit` von Sympy, da wir mit sogenannten impliziten Gleichungen arbeiten (nicht in der Form $y=f(x)$) und setzen gleich das Aspect Ratio noch auf 1:1.

```python
def plot_imp( eq ):
    p = sp.plot_implicit(eq, show=False)
    p.aspect_ratio = (1,1)
    p.show()
```

### Kreis: $x^2 + y^2 = r^2$

```python
eq = x**2 + y**2 - 1
plot_imp(eq)
```

### Ellipse: $\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$

```python
eq = x**2/2**2 + y**2/1**2 - 1
plot_imp(eq)
```

### Parabel: $y = a x^2 + b x + c$

```python
eq = y - 3*x**2 - 2*x + 1
plot_imp(eq)
```

### Hyperbel: $y = 1/x$

```python
eq = y*x - 1
plot_imp(eq)
```

## Allgemeine Quadrik

$q(\mathbf{x}) = \mathbf{x}^T \mathbf{A} \mathbf{x} + \mathbf{b}^T \mathbf{x} + c$

```python
def quadrik( A, b, c ):
    x, y = sp.symbols("x y")
    vec = sp.Matrix([[x],[y]])
    return (vec.T*A*vec + b.T*vec)[0] + c
```

### Kreis

```python
A = sp.Matrix([[1,0],[0,1]])
b = sp.Matrix([[0],[0]])
c = -1
plot_imp( quadrik( A, b, c ) )
```

### Ellipse

```python
A = sp.Matrix([[1/2,0],[0,1]])
b = sp.Matrix([[0],[0]])
c = -1
plot_imp( quadrik( A, b, c ) )
```

### Parabel

```python
A = sp.Matrix([[1,0],[0,0]])
b = sp.Matrix([[0],[-1]])
c = 0
plot_imp( quadrik( A, b, c ) )
```

### Hyperbel

```python
A = sp.Matrix([[0,0.5],[0.5,0]])
b = sp.Matrix([[0],[0]])
c = -1
plot_imp( quadrik( A, b, c ) )
```

### Probiere selbst verschiedene Werte fr A, b und c aus!

```python
A = sp.Matrix([[1,2],[3,4]])
b = sp.Matrix([[0],[0]])
c = 1
plot_imp( quadrik( A, b, c ) )
```

## Warum sind diese 4 Objekte mit der gleichen allgemeinen Formel beschreibbar?

### Fr Interessierte:

Dass die Schnittmenge einer Ebene mit einem Kegel gerade einer Quadrik entspricht, kann leicht motiviert werden.

Ein Kegel entspricht einem Kreis, dessen Durchmesser abhngig von der $z$-Achse variert:

$$x^2 + y^2 = z^2.$$

Eine Ebene in drei Dimensionen besitzt die Koordinatenform

$$\alpha x + \beta y + \gamma z + d = 0.$$

Nun knnen wir die Ebenengleichung nach $z$ auflsen und in die Kegelgleichung einsetzen:

$$x^2 + y^2 = \frac{1}{\gamma^2}(-\alpha x - \beta y - d)^2 = \frac{1}{\gamma^2} \left(\alpha^2 x^2 + \beta^2 y^2 + 2 \alpha\beta x y + 2 \alpha d x + 2 \beta d y + d^2\right),$$

und damit finden wir:

$$(\alpha^2-\gamma^2)x^2 + (\beta^2-\gamma^2)y^2 + 2\alpha\beta x y + 2 \alpha d x + 2 \beta d y + d^2 = 0.$$

Das entspricht gerade einer quadratischen Gleichung der bekannten Form:

$$\begin{pmatrix} x & y \end{pmatrix} \begin{pmatrix} \alpha^2-\gamma^2 & \alpha \beta \\ \alpha\beta & \beta^2-\gamma^2\end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} + \begin{pmatrix} 2\alpha d & 2\beta d \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} + d^2 = 0$$
